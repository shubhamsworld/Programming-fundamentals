# Aggregation (Has-A Relationship)

## What is Aggregation?

Aggregation is a relationship between two classes where one class contains or uses an object of another class.

In simple words:

> One class "Has-A" object of another class.

Examples:

* Customer has an Address
* Student has an Address
* Restaurant has a Menu
* Employee has a Department

This relationship is called **Aggregation**.

---

# Customer and Address Example

Suppose we have two classes:

```python
Customer
Address
```

A customer has an address.

Therefore:

```text
Customer Has-A Address
```

---

## Implementation

```python
class Customer:

    def __init__(self, name, gender, address):

        self.name = name
        self.gender = gender
        self.address = address
```

Notice:

```python
self.address = address
```

Here `address` is not a string.

It is actually an object of the Address class.

---

## Address Class

```python
class Address:

    def __init__(self, city, pin, state):

        self.city = city
        self.pin = pin
        self.state = state
```

Creating an Address object:

```python
Address1 = Address(
    "Bengaluru",
    560068,
    "Karnataka"
)
```

---

## Passing One Object Inside Another

```python
cust = Customer(
    "Shubham",
    "Male",
    Address1
)
```

Memory View:

```text
cust ---------> Customer Object
                  |
                  |
                  +---- address ---------> Address Object
                                              |
                                              +---- city = Bengaluru
                                              +---- pin = 560068
                                              +---- state = Karnataka
```

The Customer object contains a reference to the Address object.

This is Aggregation.

---

# Accessing Data of the Aggregated Object

```python
def print_address(self):

    print(
        self.address.city,
        self.address.pin,
        self.address.state
    )
```

Explanation:

```python
self.address
```

gives the Address object.

Then:

```python
self.address.city
```

accesses the city's value.

---

# Introducing Encapsulation

Suppose we make city private:

```python
class Address:

    def __init__(self, city, pin, state):

        self.__city = city
```

Now:

```python
self.address.__city
```

cannot be accessed directly.

Why?

Because:

```python
__city
```

is private.

---

# Getter Solves the Problem

We provide controlled access using a getter.

```python
class Address:

    def __init__(self, city, pin, state):

        self.__city = city

    def get_city(self):
        return self.__city
```

Now:

```python
self.address.get_city()
```

returns the city safely.

---

# Updated Customer Method

```python
def print_address(self):

    print(
        self.address.get_city(),
        self.address.pin,
        self.address.state
    )
```

Now the Customer class does not directly access private data.

Instead it uses the public interface provided by Address.

This is a good example of Encapsulation.

---

# Relationship Between Aggregation and Encapsulation

Aggregation:

```text
Customer Has-A Address
```

Encapsulation:

```text
Address hides its internal data
and provides controlled access
through methods.
```

Together:

```text
Customer
    |
    +---- Address Object
                |
                +---- Private Data
                |
                +---- Getter/Setter Methods
```

---

# Key Takeaways

### Aggregation

* Represents a Has-A relationship.
* One class contains or uses another class object.
* Objects can exist independently.

Example:

```text
Customer Has-A Address
```

---

### Encapsulation

* Hides internal data.
* Uses private variables.
* Provides controlled access through getters and setters.

Example:

```python

```

accessed through:

```python

```

---

# Golden Rule

Aggregation answers:

> How are classes related?

Example:

```text

```

Encapsulation answers:

> How should data inside a class be protected and accessed?

Example:


````python
class Customer:
    def __init__(self, name, gender, address):
        self.name = name
        self.gender = gender
        self.address = address

    def print_address(self):
        print(self.address.get_city(), self.address.pin, self.address.state)

    def edit_profile(self, new_name, new_city, new_pin, new_state):
        self.name = new_name
        self.address.edit_address(new_city, new_pin, new_state)


class Address:
    def __init__(self, city, pin, state):
        self.__city = city
        self.pin = pin
        self.state = state

    def get_city(self):
        return self.__city
    
    def edit_address(self, new_city, new_pin, new_state):
        self.__city = new_city
        self.pin = new_pin
        self.state = new_state


Address1 = Address('Bengaluru', 560068, 'Karnataka')
cust= Customer('Shubham', 'Male', Address1)

cust.print_address()

cust.edit_profile('Suggu', 'Mumbai', 121212, 'Maharashtra')
cust.print_address()
````

# Aggregation vs Composition

Aggregation:

```text
Customer ◇──── Address
```

* Weak ownership.
* Both objects can exist independently.
* Address can exist without Customer.

Example:

```text
Customer Has-A Address
Employee Has-A Department
Student Has-A Laptop
```

---

Composition:

```text
Customer ◆──── Address
```

(Filled Diamond)

* Strong ownership.
* Child object cannot exist without parent.
* Destroying parent destroys child.

Example:

```text
House Has-A Room
Human Has-A Heart
Car Has-A Engine
```

---

## Inheritance

DRY - Dont Repeat Yourself


````python
class User: # Parent class
    def __init__(self):
        self.name = 'Shubham'

    def login(self):
        print('login')

class Student(User): #Child class
    def __init__(self):
        self.rollno = 100 
    
    def enroll(self):
        print("Enroll into the course")

u = User()
s= Student()

print(s.name)
print(s.rollno)
s.login()
s.enroll()

# Not Running since here the method(__init__) is overrided i.e.
# Not running because the child class has overridden
# the parent's constructor (__init__).

# When Student() is created, only Student.__init__()
# runs and User.__init__() is not called automatically.

# Therefore self.name is never created.

# To initialize parent attributes as well,
# we must explicitly call:

# super().__init__()
````

````python
class User: # Parent class
    def __init__(self):
        self.name = 'Shubham'

    def login(self):
        print('login')

class Student(User): #Child class
    # def __init__(self):
    #     self.rollno = 100 
    
    def enroll(self):
        print("Enroll into the course")

u = User()
s= Student()

print(s.name)
# print(s.rollno)
s.login()
s.enroll()
````

````python
# Parent Class
class User:

    def __init__(self):

        # This attribute is created whenever
        # User's constructor runs
        self.name = 'Shubham'

    def login(self):

        # Method belonging to User class
        print('Login')


# Child Class
# Student inherits from User
class Student(User):

    def __init__(self):

        # VERY IMPORTANT
        # Calling parent's constructor

        # Without this line:
        # self.name will never be created

        super().__init__()

        # Student-specific attribute
        self.rollno = 100

    def enroll(self):

        # Method specific to Student class
        print("Enroll into the course")


# Creating Student Object
s = Student()


# -----------------------------
# What happens internally?
# -----------------------------

# 1. Student() is called
#
# 2. Python enters Student.__init__()
#
# 3. super().__init__() executes
#
# 4. User.__init__() runs
#
#       self.name = "Shubham"
#
# 5. Control returns to Student.__init__()
#
# 6. self.rollno = 100
#
# Final Student object:
#
# s
# |
# +---- name = "Shubham"    <- inherited from User
# |
# +---- rollno = 100        <- Student's own attribute
#


# Accessing inherited attribute
print(s.name)

# Accessing Student's own attribute
print(s.rollno)


# Calling inherited method
# Python searches:
#
# Student class
#      ↓
# User class
#
# Finds login() in User
s.login()


# Calling Student's own method
s.enroll()
````
