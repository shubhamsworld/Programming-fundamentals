## __add__



- __add__() is a special (dunder) method that defines what should happen when the + operator is used with objects.
- When you do:
    - print(10 + 20)
- Python internally does:
    - int'.'__add__'(10, 20)

__add__() is a special method that overloads the + operator. It defines how two objects of a class should be added together.

| Operator | Method          |
| -------- | --------------- |
| `+`      | `__add__()`     |
| `-`      | `__sub__()`     |
| `*`      | `__mul__()`     |
| `/`      | `__truediv__()` |
| `==`     | `__eq__()`      |
| `<`      | `__lt__()`      |
| `>`      | `__gt__()`      |


````python
class Fraction:

    def __init__(self, x, y): #parameterized constructor 
        self.num = x
        self.den = y

    def __str__(self):
        return '{}/{}'.format(self.num, self.den)
    
    def __add__(self, other):
        new_num = self.num*other.den + other.num*self.den
        new_den = self.den*other.den
        return '{}/{}'.format(new_num,new_den)

    def __sub__(self, other):
        new_num = self.num*other.den - other.num*self.den
        new_den = self.den*other.den
        return '{}/{}'.format(new_num,new_den)

    def __mul__(self, other):
        new_num = self.num*self.num
        new_den = other.den*other.den
        return '{}/{}'.format(new_num,new_den)
    
    def __truediv__(self, other):
        new_num = self.num*self.den
        new_den = other.den*other.num
        return '{}/{}'.format(new_num,new_den)
    
    def convert_to_decimal(self): #Not a magic method
        return self.num/self.den
         
````

````python
fr1 = Fraction(1,8)
fr2 = Fraction(7,9)
print(fr1)
print(fr1+fr2)
````

Write OOP classes to handle the following scenarios:

- A user can create and view 2D co-ordinates
- A user can find out the distance between 2 co-ordinates
- A user can find the distance of a co-ordinate from origin
- A user can check if a point lies on a given line
- A user can find the distance between  a given 2D point and a given line

````python
class Point:
    def __init__(self, x, y):
        self.x_cod = x
        self.y_cod = y

    def __str__(self):
        return '<{},{}>' .format(self.x_cod, self.y_cod)
    
    def euclidean_distance(self, other):
        return ((self.x_cod - other.x_cod)**2 + (self.y_cod - other.y_cod)**2)**0.5
    
    def distance_from_origin(self):
        #return (self.x_cod**2 + self.y_cod**2)**0.5
        return self.euclidean_distance(Point(0,0))
    
class Line:
    def __init__(self, A,B, C):
        self.A = A
        self.B = B
        self.C = C

    def __str__(self):
        return '{}x + {}y + {} = 0' .format(self.A, self.B, self.C)
    
    def point_on_line(line, point):
        if line.A * point.x_cod + line.B * point.y_cod + line.C ==0:
            return "yes, the point lies on the line"
        else:
            return "does not lie on the line"
        
    def shortest_distance(self, point):
        return abs(self.A * point.x_cod +
            self.B * point.y_cod +
            self.C) / ((self.A**2 + self.B**2) ** 0.5)
````

````python
p1 = Point(1,1)
p2 = Point(10,10)


print(p1)
p1.euclidean_distance(p2)
p1.distance_from_origin()

l1= Line(1,1,-2)
print(l1)

l1.point_on_line(p1)
l1.shortest_distance(p1)
````

## Pass by reference

In Python, objects can be:

Passed as arguments to functions.
Returned from functions.

A function receives a reference to the object, not a copy of the object.

````python
class Person:

    def __init__(self, name, gender):
        self.name=name
        self.gender=gender
        
# outside class and hence its a function, not a method
def greet(person): # we are giving object as input to a function

    print('Hi, my name is', person.name,'and i am a', person.gender)

# A function can also return a object of a class
    p1 = Person('suggu', 'male')
    return p1
p = Person('Shubham', 'male')
o = greet(p)
print(o.name)
````

Passing Objects to Functions
Objects can be passed as arguments to functions.
The function receives a reference to the same object.
No copy of the object is created.
Changes made to the object's attributes inside the function are visible outside the function.

Variables do not store objects. Variables store references to objects. Multiple variables can refer to the same object, and changes through one reference are visible through all others.

````python
class Person:

    def __init__(self, name, gender):
        self.name=name
        self.gender=gender
        
# outside class and hence its a function, not a method
def greet(person): # we are giving object as input to a function
    print(id(person))
    person.name= 'Suggu' #The function is not changing the class. It is changing the object that was passed to it.
    print(person.name)

P= Person('Shubham', 'Male')
print(id(P))
greet(P)
print(P.name)
````

## Mutability of the object


All the objects of user defined classes in python are mutable( though we can make them immutable, as well)

Memory address changes when a new object is created.

Mutability determines whether Python can modify the existing object or needs to create a new object.

The address remains the same because the function returns a reference to the same object. Modifying an attribute changes the existing object rather than creating a new one. A different address is obtained only when a new object is created and returned.

````python
class Person:

    def __init__(self, name, gender):
        self.name=name
        self.gender=gender

def greet(person): # we are giving object as input to a function
    person.name = 'Suggu'
    return person

p = Person('Shubham', 'Male')
print(id(p))
p1 = greet(p)
print(id(p1)) # if the object of the class is mutable, then address wont change but if the object of the class is immutable then the address will be changed.
````

## Encapsulation and what problem does it solve


Encapsulation is the OOP principle of combining data (attributes) and the methods that operate on that data into a single unit called a class, while controlling how the data is accessed and modified.

To achieve encapsulation, we make important attributes private using double underscores (`__`). Python performs name mangling on these attributes, making direct access difficult and reducing accidental misuse.

Since private data should not be accessed directly, we provide public methods such as getters and setters (or other business methods like `deposit()`, `withdraw()`, etc.) to interact with that data. These methods act as controlled gateways that can validate input, enforce rules, and protect the object's internal state.

Thus:

* Private Variables → Hide internal data
* Getters → Controlled read access
* Setters → Controlled write access
* Public Methods → Safe interaction with the object

Encapsulation helps maintain data integrity, improves security, reduces unintended modifications, and makes classes easier to maintain and use correctly.

**In simple words:** Encapsulation means "hide the implementation details and expose only what is necessary through a controlled interface."


```text
Encapsulation
       |
       |
       +---- Private Variables
       |          |
       |          +---- Hide Internal Data
       |
       +---- Getters
       |          |
       |          +---- Controlled Read Access
       |
       +---- Setters
                  |
                  +---- Controlled Write Access
```

# Private Variables and Private Methods in Python

## 1. Why Do We Need Private Members?

In OOP, some data and methods are meant to be used **only inside the class**.

For example:

* ATM balance
* ATM PIN
* Database credentials
* Internal helper methods

We don't want users of the class to modify them directly.

---

## 2. Public Variables

By default, every variable in Python is public.

```python
class Atm:
    def __init__(self):
        self.balance = 0
```

Usage:

```python
obj = Atm()

obj.balance = 10000
print(obj.balance)
```

Output:

```text
10000
```

Anyone can access and modify it.

---

## 3. Protected Variables (`_variable`)

```python
class Atm:
    def __init__(self):
        self._balance = 0
```

The single underscore is only a convention.

It means:

> "This variable is intended for internal use. Please don't access it directly."

But Python still allows:

```python
obj._balance = 10000
```

No error occurs.

---

## 4. Private Variables (`__variable`)

```python
class Atm:
    def __init__(self):
        self.__balance = 0
```

Double underscore tells Python:

> "Make accidental access difficult."

This is called **Name Mangling**.

---

## 5. What is Name Mangling?

Python changes:

```python
self.__balance
```

to:

```python
self._Atm__balance
```

internally.

So:

```python
class Atm:
    def __init__(self):
        self.__balance = 1000

obj = Atm()
```

Object actually stores:

```python
obj._Atm__balance
```

---

## 6. Why Does This Fail?

```python
obj.__balance
```

Error:

```text
AttributeError
```

Because Python looks for:

```python
__balance
```

which does not exist.

The actual name is:

```python
_Atm__balance
```

---

## 7. Can We Still Access It?

Yes.

```python
obj._Atm__balance
```

Output:

```text
1000
```

Python does not provide true privacy.

---

## 8. Why Use Private Variables Then?

To prevent accidental misuse.

Example:

Without private variables:

```python
obj.balance = -100000
```

Anyone can corrupt the data.

Instead:

```python
class Atm:
    def __init__(self):
        self.__balance = 0
```

Users are encouraged to use methods:

```python
deposit()
withdraw()
check_balance()
```

instead of modifying balance directly.

---

## 9.  Important Observation

Suppose:

```python
class Atm:
    def __init__(self):
        self.__balance = 0

obj = Atm()
```

Now:

```python
obj.balance = "hehehe"
```

Because Python creates a NEW attribute:


These are two different attributes.

## 10. Private Methods

Just like variables:

```python

    def __menu(self):
        print("Private Menu")
```

Python converts:

```python

```

to:

```python
_Atm__menu
```

This fails:

```python
obj.__menu()
```

Error:


Because method name has been mangled.

Public method calls private method.

## 13. Why Private Methods?

Sometimes a method is only a helper.

Example:

```python
class Bank:

        pass

```

User only needs:

```python
withdraw()
```

not:

```python

```

## Getter and Setter Methods

- for every data that we have in our class, we make two methods 
    - setter
    - getter
    

## Why Do We Need Getters and Setters?

When we make an attribute private using double underscores (`__`), it should not be accessed directly from outside the class.

```python
class Atm:
    def __init__(self):
        self.__balance = 0
```

Now:

```python
obj.__balance
```

will raise an error.

To provide controlled access to private data, we use **Getters** and **Setters**.

---

## Getter

A getter is a method used to **read** the value of a private attribute.

Example:

```python
class Atm:

    def __init__(self):
        self.__balance = 0

    def get_balance(self):
        return self.__balance
```

Usage:

```python
obj.get_balance()
```

### Purpose of Getter

* Provides controlled read access to private data.
* Allows users to view data without directly accessing the variable.
* Maintains encapsulation.

---

## Setter

A setter is a method used to **modify** the value of a private attribute.

Example:

```python
class Atm:

    def __init__(self):
        self.__balance = 0

    def set_balance(self, new_value):
        self.__balance = new_value
```

Usage:

```python
obj.set_balance(1000)
```

### Purpose of Setter

* Provides controlled write access to private data.
* Allows validation before updating data.
* Prevents invalid values from being stored.

Example with validation:

```python
def set_balance(self, new_value):

    if isinstance(new_value, int):
        self.__balance = new_value
    else:
        print("Please enter an integer value.")
```

Now:

```python
obj.set_balance("hehehe")
```

will not update the balance.

---

## How Getters and Setters Support Encapsulation

Without getters and setters:

```python
obj.balance = -100000
```

Anyone can modify the data directly.

With encapsulation:

```python
obj.set_balance(1000)
print(obj.get_balance())
```

All access goes through methods controlled by the class.

This allows the class to protect its internal state and enforce business rules.

---

## Summary

* Getter → Used to read private data.
* Setter → Used to modify private data.
* Getter provides controlled read access.
* Setter provides controlled write access.
* Together they help implement encapsulation by protecting and controlling access to an object's internal data.

### Golden Rule

Getters and setters are methods that act as controlled gateways to private variables, allowing safe access and modification while preserving encapsulation.



````python
class Atm:

    def __init__(self):
        self.__pin = ''
        self.__balance = 0

    # Getter
    def get_balance(self):
        return self.__balance

    # Setter
    def set_balance(self, new_value):
        if isinstance(new_value, int):
            self.__balance = new_value
        else:
            print("Please enter an integer value.")

    def menu(self):

        user_input = input("""
Hi, How can I help you today?

1. Press 1 to Create Pin
2. Press 2 to Change Pin
3. Press 3 to Check Balance
4. Press 4 to Withdraw
5. Anything else to Exit

""")

        if user_input == '1':
            self.create_pin()

        elif user_input == '2':
            self.change_pin()

        elif user_input == '3':
            self.check_balance()

        elif user_input == '4':
            self.withdraw()

        else:
            print("Thank you for using ATM!")
            return

    def create_pin(self):

        user_pin = input("Enter your PIN: ")
        self.__pin = user_pin

        user_balance = int(input("Enter initial balance: "))
        self.__balance = user_balance

        print("PIN created successfully.")
        self.menu()

    def change_pin(self):

        old_pin = input("Enter old PIN: ")

        if old_pin == self.__pin:

            new_pin = input("Enter new PIN: ")
            self.__pin = new_pin

            print("PIN changed successfully.")

        else:
            print("Wrong PIN.")

        self.menu()

    def check_balance(self):

        user_pin = input("Enter your PIN: ")

        if user_pin == self.__pin:
            print("Your balance is:", self.__balance)

        else:
            print("Wrong PIN.")

        self.menu()

    def withdraw(self):

        user_pin = input("Enter your PIN: ")

        if user_pin == self.__pin:

            amount = int(input("Enter amount to withdraw: "))

            if amount <= self.__balance:

                self.__balance -= amount

                print(
                    "Withdrawal successful."
                )
                print(
                    "Remaining balance:",
                    self.__balance
                )

            else:
                print("Insufficient balance.")

        else:
            print("Wrong PIN.")

        self.menu()
````

# Static Variables vs Instance Variables

## Introduction

A class can contain two types of variables:

1. Instance Variables
2. Static Variables (Class Variables)

The main difference is **who owns the variable**.

* Instance Variables belong to individual objects.
* Static Variables belong to the class itself and are shared by all objects.

---

# 1. Instance Variables

Instance variables are variables whose values are different for different objects.

They are created using `self` and are usually initialized inside the constructor.

```python
class Person:

    def __init__(self, name, age):
        self.name = name
        self.age = age
```

Here:

```python
self.name
self.age
```

are instance variables.

---

## Characteristics of Instance Variables

* Belong to an object.
* Created using `self`.
* Stored inside the object.
* Each object gets its own copy.
* Changing one object's value does not affect another object.

---

## Example

```python
class Person:

    def __init__(self, name):
        self.name = name

p1 = Person("Shiva")
p2 = Person("Nandani")

p1.name = "Suggu"

print(p1.name)
print(p2.name)
```

Output:

```text
Suggu
Nandani
```

Each object stores its own data.

Memory View:

```text
p1 -----> Person Object
            name = Suggu

p2 -----> Person Object
            name = Nandani
```

---

# 2. Static Variables (Class Variables)

Static variables belong to the class itself.

They are declared directly inside the class but outside any method.

```python
class Person:

    country = "India"
```

Here:

```python
country
```

is a static variable.

---

## Characteristics of Static Variables

* Belong to the class.
* Shared by all objects.
* Only one copy exists in memory.
* Stored inside the class.
* Changes are visible to all objects.

---

## Example

```python
class Person:

    country = "India"

    def __init__(self, name):
        self.name = name

p1 = Person("Shiva")
p2 = Person("Nandani")

print(p1.country)
print(p2.country)
```

Output:

```text
India
India
```

---

## Changing a Static Variable

Preferred way:

```python
Person.country = "USA"
```

Now:

```python
print(p1.country)
print(p2.country)
```

Output:

```text
USA
USA
```

All objects see the change because they share the same variable.

---

# ATM Example (Most Important)

Suppose we are building an ATM system.

Every customer has:

* Customer ID
* PIN
* Balance

Since every customer has different values, these should be Instance Variables.

```python
class Atm:

    def __init__(self, customer_id, pin, balance):

        self.customer_id = customer_id
        self.pin = pin
        self.balance = balance
```

Example:

```python
c1 = Atm(101, "1234", 5000)
c2 = Atm(102, "5678", 10000)
```

Memory:

```text
c1 -----> Atm Object
            customer_id = 101
            pin = 1234
            balance = 5000

c2 -----> Atm Object
            customer_id = 102
            pin = 5678
            balance = 10000
```

These are Instance Variables because every customer has unique values.

# Static Variable as a Counter

One of the most common uses of static variables is keeping track of information shared by all objects.

Suppose the bank wants to know:

```text
How many ATM accounts have been created?
```

This information belongs to the class, not to any specific customer.

```python
class Atm:

    total_customers = 0

    def __init__(self, customer_id, pin, balance):

        self.customer_id = customer_id
        self.pin = pin
        self.balance = balance

        Atm.total_customers += 1
```

Usage:

```python
c1 = Atm(101, "1234", 5000)
c2 = Atm(102, "5678", 10000)
c3 = Atm(103, "9999", 15000)

print(Atm.total_customers)
```

Output:

```text
3
```

Since all objects share the same counter, it is a Static Variable.

# Instance Variable vs Static Variable

| Feature             | Instance Variable       | Static Variable       |
| ------------------- | ----------------------- | --------------------- |
| Belongs To          | Object                  | Class                 |
| Created Using       | self.variable           | variable inside class |
| Copies              | One per object          | One shared copy       |
| Memory Location     | Object                  | Class                 |
| Access              | obj.variable            | Class.variable        |
| Modification Effect | Affects only one object | Affects all objects   |

# Best Practice

For Static Variables:

```python
ClassName.variable
```

Preferred:

```python
Atm.total_customers
```

instead of:

```python
obj.total_customers
```

because the variable belongs to the class.

---

# Golden Rule

Instance Variables store data that is unique to each object.

Static Variables store data that is shared by all objects of a class.

If every object needs its own copy → Use an Instance Variable.

If all objects should share the same value → Use a Static Variable.
