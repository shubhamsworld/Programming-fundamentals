## Constructor 
- its a method, (as its a function defined inside a class) but its a Magic method and have a super power (that its automatically called whenever a object of the class is built).
- A constructor is a special method that is executed automatically when an object of a class is created.
### But whats the benefit of Constructor 
- Normally, the user decides which method to call and when to call it.
- However, some tasks are essential and should happen automatically whenever an object is created. Such tasks are written inside the constructor.
    - We write code in a constructor for:

        - Initial setup of the object
        - Configuration-related tasks
        - Initializing variables
        - Establishing required connections
- The user does not need to call these operations manually because they run automatically when the object is created.
- Examples
    - Connecting the application to a database
    - Loading configuration settings
    - Connecting to an API or internet service
    - Initializing important variables

## Self
- self is a reference to the current object of a class.

    - Whenever an object calls a method, Python automatically passes that object as the first argument to the method, which is usually named self.

- For every instance method in a class, the first parameter should refer to the current object and is conventionally written as self.

### Why do we use self?

# self allows an object to:

Access its attributes (variables)
Access its methods
Distinguish its own data from other objects' data

Without self, Python would not know which object's name to use.

### Golden Rule of OOP
An object can access its own:

- Attributes (data members)
- Methods (member functions)

using self.

Can one method call another method of the same class?

Yes. It is done using self. 
But not possible without self

```python
class Student: 
    def show(self):
        print("Hello")

s1 = Student()
s1.show()

# Python internally does:
# Student.show(s1)
# Here, self refers to s1./// 
```

```python
class Atm:

    # constructor (Constructor is a special function)-> Dont need to call it explicitly, whenever object is created, constructor fun automatically runs 
    def __init__(self):
        self.pin =''
        self.balance = 0
        print(id(self))

    def menu(self):
        user_input = input(""" Hi, How can I help you today?
              1. Press 1 to create Pin
              2. Press 2 to change Pin
              3. Press 3 to check Balance
              4. Press 4 to Withdraw
              5. Anything else to Exit
""")
        if user_input == '1':
            # create pin
            self.create_pin()
        elif user_input == '2':
            # change pin
            self.change_pin()
        elif user_input == '3':
            # check balance
            self.check_balance()
        elif user_input == '4':
            # withdraw
            self.withdraw()
        else:
            exit()
        


    def create_pin(self):
        user_pin = input("Enter your pin")
        self.pin = user_pin

        user_balance = int(input("Enter balance"))
        self.balance = user_balance
        print("Pin created successfully")
        self.menu()

    def change_pin(self):
        old_pin = input("Enter old pin")
        if old_pin == self.pin:
            # let him change the pin
            new_pin = input("Enter new pin")
            self.pin = new_pin
            print("Pin changed successfully")
            self.menu()
        else:
            print("You entered wrong pin")
            self.menu()

    def check_balance(self):
        user_pin = input("Enter your pin")
        if user_pin == self.pin:
            print("Your balance is", self.balance)
        else:
            print("You entered wrong pin")
        self.menu()

    def withdraw(self):
        user_pin = input("Enter your pin")
        if user_pin == self.pin:
            amount = int(input("Enter the amount"))
            if amount <= self.balance:
                self.balance = self.balance - amount
                print("Withdrawl successful, remaining balance is,",self.balance)
            else:
                print("Insufficient balance")
        else:
            print("You entered wrong pin")
        self.menu()




obj = Atm()

print(id(obj))
```

## __str__
- This also is a Magic method (or Dunder method).
- __str__() is a special method (dunder method) that controls what is displayed when you print an object.

### When is __str__() called?

- It is automatically called when:
    - print(obj)
    - str(obj)
    - f"{obj}"
are used.

### Why do we use it?

- To provide a human-readable representation of an object.

- __str__() is a special method used to define the string representation of an object. It is automatically called when the object is printed or converted to a string.


## __add__
- __add__() is a special (dunder) method that defines what should happen when the + operator is used with objects.
- When you do:
    - print(10 + 20)
- Python internally does:
    - int.__add__(10, 20)

__add__() is a special method that overloads the + operator. It defines how two objects of a class should be added together.

| Operator | Method          |
| -------- | --------------- |
| `+`      | `__add__()`     |
| `-`      | `__sub__()`     |
| `*`      | `__mul__()`     |
| `/`      | `__truediv__()` |
| `==`     | `__eq__()`      |
| `<`      | `__lt__()`      |
| `>`      | `__gt__()`      |


```python
class Fraction:

    def __init__(self, x, y): #parameterized constructor 
        self.num = x
        self.den = y

    def __str__(self):
        return '{}/{}'.format(self.num, self.den)
    
    def __add__(self, other):
        new_num = self.num*other.den + other.num*self.den
        new_den = self.den*other.den
        return '{}/{}'.format(new_num,new_den)

    def __sub__(self, other):
        new_num = self.num*other.den - other.num*self.den
        new_den = self.den*other.den
        return '{}/{}'.format(new_num,new_den)

    def __mul__(self, other):
        new_num = self.num*self.num
        new_den = other.den*other.den
        return '{}/{}'.format(new_num,new_den)
    
    def __truediv__(self, other):
        new_num = self.num*self.den
        new_den = other.den*other.num
        return '{}/{}'.format(new_num,new_den)
    
    def convert_to_decimal(self): #Not a magic method
        return self.num/self.den
         
```

```python
fr1 = Fraction(1,8)
fr2 = Fraction(7,9)
print(fr1)
print(fr1+fr2)
```
